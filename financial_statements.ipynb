```python
def pull_basic_data(function, symbol, freq = 'quarterlyReports'):
    # func_to_key = {'INCOME_STATEMENT':'quarterlyReports', 'BALANCE_SHEET':'quarterlyReports','CASH_FLOW':'quarterlyReports'}
    url = f'https://www.alphavantage.co/query?function={function}&symbol={symbol}&apikey={api_keys.get_api_key()}'
    r = requests.get(url)
    data = r.json()
    
    df = pd.DataFrame.from_records(data[freq])
    return df
```

```python
dfs = {}
symbols = ['AAPL']
for symbol in symbols:
    if symbol not in dfs.keys():
        dfs[symbol] = {}
    for function in key_financial_statements:
        dfs[symbol][function] = pull_basic_data(function,symbol)
```

```python
from functools import cache
from typing import List, Optional
from collections import Counter, defaultdict
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        # @cache
        def dp(s1,s2,i,j):
            if i==len(s1) or j==len(s2): return 0
            if s1[i]==s2[j]: return 1+dp(s1,s2,i+1,j+1)
            return max(dp(s1,s2,i+1,j),dp(s1,s2,i,j+1))
        return dp(text1,text2,0,0)
```
